## Examples of usage

```python
import numpy as np
import random
import pandas as pd
import meliora.meliora.val_tests as vt
```

```python
# Test Jeffrey's test
test_data = pd.read_csv('pd_test_data.csv')
result= vt.jeffreys_test(test_data, 'ratings', 'default_flag', 'predicted_pd', 0.05)
result
```

```python
# Jeffrey's test
vt.jeffreys_test(test_data, "ratings", "default_flag", "predicted_pd", alpha_level=0.1)
```

```python
# Binomial test
vt.binomial_test(test_data, "ratings", "default_flag", "predicted_pd")
```

```python
# Brier score
vt.brier_score(test_data, "ratings", "default_flag", "predicted_pd")
```

```python
# Herfindahl test
buckets = ['A', 'B', 'C']
bucket_pds = {'A': .1, 'B': .15, 'C': .2}
ratings2 = random.choices(buckets,  [0.4, 0.5, 0.1], k=1000)
probs_default2 = [bucket_pds[rating] for rating in ratings2]
default_flag2 = [random.uniform(0, 1) < bucket_pds[rating]
                for rating in ratings2]
test_data2 = pd.DataFrame({'ratings': ratings2, 
                         'default_flag': [int(i) for i in default_flag2], 
                         'predicted_pd' : probs_default2})

test_data2.to_csv('pd_test_data2.csv', index=False)

vt.herfindahl_test(test_data, test_data2, "ratings")
```

```python
# Hosmer-Lemeshow test
vt.hosmer_test(test_data, "ratings", "default_flag", "predicted_pd")
```

```python
# Spiegelhalter test
vt.spiegelhalter_test(test_data, "ratings", "default_flag", "predicted_pd")
```
